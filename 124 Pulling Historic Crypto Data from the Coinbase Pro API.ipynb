# List of Cryptos

```python
from Historic_Crypto import Cryptocurrencies, HistoricalData
import pandas as pd 
import cryptocompare
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
data = Cryptocurrencies(coin_search = 'XLM', extended_output=False).find_crypto_pairs()
cryptocompare.get_coin_list(format=False)
```

# 1 Pulling Historic Crypto

```python
frequency = 86400 #daily price data

# top 10 coins
coinlist = ['BTC-USD', 'LTC-USD', 'BCH-USD', 'XLM-USD', 'LINK-USD'] 
for i in range(len(coinlist)):
    coinname = coinlist[i]
    pricedata = HistoricalData(coinname,frequency,'2010-01-01-00-00').retrieve_data()
    pricedf = pricedata['close']
    
    if i == 0:
        df = pd.DataFrame(pricedf.copy())
    else:
        df = pd.merge(left=df, right=pricedf, how='left', left_index=True, right_index=True) 
    
    df.rename(columns={"close": coinname}, inplace=True)
df 
```

# 2 Price Chart

```python
from datetime import date, timedelta, datetime
# This function adds plotting functions for calender dates
from pandas.plotting import register_matplotlib_converters
# Important package for visualization - we use this to plot the market data
import matplotlib.pyplot as plt 

# Visualize the data
fig, ax1 = plt.subplots(figsize=(22, 10), sharex=False)

for i in range(len(coinlist)):
    coinname = coinlist[i]
    x = df.index
    y = df[coinname] 
    plt.ylabel("Dollar", fontsize=18)
    plt.plot(x, y, linewidth=2.0)

plt.title("Crypto Prices", fontsize=20)
plt.legend(coinlist, loc="upper left")

plt.grid()
plt.show()
```

# 3 Multi Chart

```python
# Percentage Gains Chart
scaler = MinMaxScaler(feature_range=(0, 1))
scaler = scaler.fit(df[coinlist])
dfp = scaler.transform(df[coinlist])
dfp = pd.DataFrame(dfp, columns = coinlist, index = df.index)

# Visualize the data
fig, ax1 = plt.subplots(figsize=(22, 10), sharex=False)

for i in range(len(coinlist)):
    coinname = coinlist[i]
    x = dfp.index
    y = dfp[coinname] 
    plt.ylabel("Dollar", fontsize=18)
    plt.plot(x, y, linewidth=2)

plt.title("Crypto Prices", fontsize=20)
plt.legend(coinlist, loc="upper left")
plt.grid()
plt.show()
```
